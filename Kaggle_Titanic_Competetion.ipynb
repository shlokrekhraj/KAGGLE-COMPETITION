```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv("train.csv")
test_df  = pd.read_csv("test.csv")
```

```python
train_df
```

```python
#concating the dataset to find identical features for feature engineering
df = pd.concat([train_df, test_df], ignore_index=True)
df
```

```python
# Extract Mr, Mrs, Miss, Master etc and create another column with these titles
#expand=False,    Tells pandas to return a Series instead of a DataFrame.
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
# Merge rare titles into one group
df['Title'] = df['Title'].replace(['Lady','Countess','Capt','Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona'], 'Rare')
df['Title'] = df['Title'].replace(['Mlle','Ms'], 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
df
```

```python
# SibSp	    # of siblings / spouses onboard
# Parch	    # of parents / children onboard
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 #Why + 1? ,Because you must count the passenger themself.
```

```python
df['IsAlone'] = (df['FamilySize'] == 1).astype(int) #this means whose famsize==1 they r alone
df['IsAlone']
```

```python
#Fare per person (true wealth signal)
df['FarePerPerson'] = df['Fare'] / df['FamilySize']
df['FarePerPerson']
```

```python
#Only richer passengers had cabins recorded.
df['CabinHas'] = df['Cabin'].notnull().astype(int)
df['CabinHas']
```

```python
# People sharing same ticket likely traveled together
df['TicketGroup'] = df.groupby('Ticket')['Ticket'].transform('count')
df['TicketGroup']
```

```python
# Handle Missing Values
df['Age'] = df['Age'].fillna(df.groupby('Title')['Age'].transform('median'))

df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
# Age Binning
#Survival isn’t linear with age — children and young women survived much more than middle-aged men. So binning lets the model learn step-like survival patterns instead of smooth curves.
df['AgeBin'] = pd.cut(df['Age'], bins=[0,12,20,35,55,80]).astype(str)
df['AgeBin']
```

```python
# Survival is NOT just sex or class — it is the combination:
df['SexPclass'] = df['Sex'] + "_" + df['Pclass'].astype(str) # sex x Pclass  eg:-rich_men/ poor_women
df['EmbarkPclass'] = df['Embarked'] + "_" + df['Pclass'].astype(str) #embark x Pclass  eg:- S_1/C_3 etc
```

```python
df
```

## TRAIN TEST SPLIT

```python
# this gets back to the reverse of concatenation we did at start so that we train and test on diff dataset that was given by Kaggle
train2_df = df.iloc[:len(train_df)]
test2_df = df.iloc[len(train_df):]

#from features we removed "survived" to only use it as o/p and not i/p
features = [
 'Pclass','Sex','Age','Fare','Embarked','Title','FamilySize','IsAlone',
 'FarePerPerson','CabinHas','TicketGroup','AgeBin','SexPclass','EmbarkPclass'
]

X = train2_df[features]
y = train2_df['Survived']

X_test = test2_df[features]
```

# CATBOOST MODEL

```python
!pip install catboost
from catboost import CatBoostClassifier

cat_cols = X.select_dtypes(include=['object','category']).columns.tolist()

model = CatBoostClassifier(
    iterations=1200,
    depth=7,
    learning_rate=0.03,
    loss_function='Logloss',
    eval_metric='Accuracy',
    verbose=200
)

model.fit(X, y, cat_features=cat_cols)
```

```python
pred = model.predict(X_test)

submission = pd.DataFrame({
    'PassengerId': test_df['PassengerId'],
    'Survived': pred.astype(int)
})

submission.to_csv("submission.csv", index=False)
```
